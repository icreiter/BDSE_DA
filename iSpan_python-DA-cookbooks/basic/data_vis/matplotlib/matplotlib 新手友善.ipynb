## 資料準備


```python
X = np.arange(0, 12.1, 0.1)
Y = np.sin(X)

print(X)
print(len(X))
print(Y)
print(len(Y))
```

## 字體配置

```python
# 坐標軸標題，圖表標題
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

ax = plt.gca() #gca表產生座標軸的物件

ax.set_title("sin(x) function")
ax.set_xlabel("time(sec)")
ax.set_ylabel("Intensity (W)")
```

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

# 把軸取出來
ax = plt.gca()

# 粗斜體
ax.set_title("sin(x) function", fontname = "Arial", fontsize=20, weight="bold", style="italic") # 針對title來進行更改
ax.set_xlabel("Time (s)", fontname = "Arial", fontsize=13)
ax.set_ylabel("Intensity (W)", fontname = "Arial", fontsize=13)
```

## 坐標軸刻度配置

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

ax = plt.gca()

ax.set_title("sin(x) function", fontname = "Arial", fontsize=20)
ax.set_xlabel("Time (s)", fontname = "Arial", fontsize=13)
ax.set_ylabel("Intensity (W)", fontname = "Arial", fontsize=13)

ax.set_xticks([0, 2.5, 5, 7.5, 10, 12.5]) # tick指每個軸距的距離多少
```

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

ax = plt.gca()

ax.set_title("sin(x) function", fontname = "Arial", fontsize=20)
ax.set_xlabel("Time (s)", fontname = "Arial", fontsize=13)
ax.set_ylabel("Intensity (W)", fontname = "Arial", fontsize=13)

# 換標籤座標軸
_ = ax.set_xticklabels(["A", "B", "C", "D", "E", "F", "G", "H"])
```

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, color="lime", linestyle="-", linewidth=2, \
         marker="*", markerfacecolor= "red", markeredgecolor="blue" )

ax = plt.gca()

ax.set_title("sin(x) function", fontname = "Arial", fontsize=20)
ax.set_xlabel("Time (s)", fontname = "Arial", fontsize=13)
ax.set_ylabel("Intensity (W)", fontname = "Arial", fontsize=13)

ax.set_xticklabels(["A", "B", "C", "D", "E", "F", "G", "H"])
# 座標軸刻度線調整
ax.tick_params(axis="both", direction = "in", color="orange", length=10, width=3)
```

## 多線圖

```python
plt.plot(X,Y)
plt.plot(X+3,Y)
```

### 設定標籤

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, linestyle="solid", linewidth=5, label="noraml")
plt.plot(X+3,Y, linestyle="--", linewidth=5, label="shift 3 time period")

ax = plt.gca()

ax.set_title("sin(x) function")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Intensity (W)")

plt.legend() # legend 表示圖例，用以區分線條
```

```python
# 坐標軸標題，圖表標題，字形配置
plt.plot(X,Y, linestyle="solid", linewidth=5, label="noraml")
plt.plot(X+3,Y, linestyle="--", linewidth=5, label="shift 3 time period")

ax = plt.gca()

ax.set_title("sin(x) function")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Intensity (W)")

plt.legend(loc = "lower left") # 移動圖例
```

```python
# zorder 圖層順序
plt.plot(X,Y, linestyle="solid", linewidth=5, label="noraml", zorder=10)
plt.plot(X+3,Y, linestyle="--", linewidth=5, label="shift 3 time period", zorder=0)
plt.plot(X+6,Y, linestyle="--", linewidth=5, label="shift 3 time period", zorder=0)

ax = plt.gca()

ax.set_title("sin(x) function")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Intensity (W)")

plt.legend(loc = "lower left")

```

## 子圖

```python
# 開畫布 (建一個空白的框框)
fig, ax = plt.subplots(2,1) # 2 row + 1 column。 fig用來繪製
```

```python
# 開畫布 2x1
fig, ax = plt.subplots(2,1)

ax[0].plot(X,Y)
ax[1].plot(X,Y, 'r')
```

```python
# 開畫布 3x2
fig, ax = plt.subplots(3,2)
ax[2,1].plot(X,Y)

ax[1,0].plot(X+3,Y)
# ax[1,0].set_xlim([0,1])
```

```python
fig, ax = plt.subplots(2,1)

ax[0].plot(X,Y)
ax[1].plot(X+3,Y, 'r')

# 配置座標軸範圍
ax[0].set_xlim([0,10])
ax[1].set_xlim([0,10])

```

```python
fig, ax = plt.subplots(2,1)

x = np.arange(1,10)
y = np.exp(x)

ax[0].plot(x,y)
ax[1].plot(x,y, 'r')

# log座標變換
ax[1].set_yscale('log') 
```

### 雙y軸

```python
x = np.arange(1,10)
y = np.exp(x)

ax = plt.gca()

plt.plot(x,y)
# plt.plot(x, y*100)

# 共用x軸
ax2 = ax.twinx() # 雙Y軸等於共用一個X軸

# ax2.plot(x+3,y/10,"r")
ax2.plot(x+3,y*100,"r")
```

### 雙x軸

```python
x = np.arange(1,10)
y = np.exp(x)

ax = plt.gca()

plt.plot(x,y)

# 共用y軸
ax2 = ax.twiny()

ax2.plot(x+3,y/10,"r")
```

## 輸出圖表

```python
# 改變畫布大小
fig = plt.figure(figsize=(8,4))

# 建立假數據
x = np.arange(1,10)
y = np.exp(x)

# 取出軸資訊
ax = plt.gca()

# 畫圖
plt.plot(x,y)

# 共用y軸
ax2 = ax.twiny()

ax2.plot(x+3,y/10,"r")

# 存儲圖片, 設定解析度
plt.savefig('./view.png', dpi=500) # 解析度只能用此指令來打
```
